```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense, Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import os
#from datetime import datetime
import datetime
from jupyterthemes import jtplot
jtplot.style()
import pandas_datareader as web
import math
from sklearn.metrics import mean_squared_error
```

```python
list_code = ['BTC','ETH','BNB','XRP','DOGE','ADA','BCH','LTC','SOL','WIN','LINK']

start = datetime.datetime(2020,1,1)
end = datetime.datetime.now()
```

```python
for code in list_code:
    os.chdir(f'C:/Users/luc57.DESKTOP-NB5DC80/Side Projects/Time Series/')
    if not os.path.exists(code):
        os.makedirs(code)
    os.chdir(f'C:/Users/luc57.DESKTOP-NB5DC80/Side Projects/Time Series/{code}/')
    crypto_currency = code
    against_currency = 'USD'
    
    try:
        df = web.DataReader(f'{crypto_currency}-{against_currency}','yahoo',start,end)
        df = df.reset_index()
        trainX = []
        trainY = []

        n_future = 1   # Number of days we want to predict into the future
        n_past = 14     # Number of past days we want to use to predict the future

        #Separate dates for future plotting
        train_dates = pd.to_datetime(df['Date'])

        #Variables for training
        cols = list(df)[1:6]

        df_for_training = df[cols].astype(float)

        scaler = StandardScaler()
        scaler = scaler.fit(df_for_training)
        df_for_training_scaled = scaler.transform(df_for_training)

        for i in range(n_past, len(df_for_training_scaled) - n_future +1):
            trainX.append(df_for_training_scaled[i - n_past:i, 0:df_for_training.shape[1]])
            trainY.append(df_for_training_scaled[i + n_future - 1:i + n_future, 0])

        trainX, trainY = np.array(trainX), np.array(trainY)

        model = Sequential()
        model.add(LSTM(64, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
        model.add(LSTM(32, activation='relu', return_sequences=False))
        model.add(Dropout(0.2))
        model.add(Dense(trainY.shape[1]))

        model.compile(optimizer='adam', loss='mse')
        model.summary()

        # fit model
        history = model.fit(trainX, trainY, epochs=10, batch_size=16, validation_split=0.1, verbose=1)

        n_future=120  #Redefining n_future to extend prediction dates beyond original n_future dates...

        forecast_period_dates = pd.date_range(list(train_dates)[-100], periods=n_future, freq='1d').tolist()

        forecast = model.predict(trainX[-n_future:]) #forecast 

        #Perform inverse transformation to rescale back to original range
        #Since we used 5 variables for transform, the inverse expects same dimensions
        #Therefore, let us copy our values 5 times and discard them after inverse transform
        forecast_copies = np.repeat(forecast, df_for_training.shape[1], axis=-1)
        y_pred_future = scaler.inverse_transform(forecast_copies)[:,0]

        forecast_dates = []
        for time_i in forecast_period_dates:
            forecast_dates.append(time_i.date())

        df_forecast = pd.DataFrame({'Date':np.array(forecast_dates), 'Open':y_pred_future})
        df_forecast['Date']=pd.to_datetime(df_forecast['Date'])


        original = df[['Date', 'Open']]
        original['Date']=pd.to_datetime(original['Date'])
        original = original.loc[original['Date'] >= '2020-5-1']

        sns.lineplot(original['Date'], original['Open'])
        sns.lineplot(df_forecast['Date'], df_forecast['Open'])
        plt.savefig(f"output_{code}.png")
        plt.cla()

        df_forecast=df_forecast[(df_forecast['Date']>='2021-02-01')]
        df_forecast.rename(columns={'Open':'forecast_open'},inplace=True)

        df_to_compare_actual=df[(df['Date']>='2021-02-01')].reset_index()
        df_to_compare_actual_open = df_to_compare_actual[['Date','Open']]
        df_comparison = df_to_compare_actual_open.rename(columns={'Open':'actual_open'})
        df_comparison = df_comparison.merge(df_forecast, how='left', on='Date')
        df_comparison.to_excel(f'df_{code}.xlsx')

        #calculate RMSE
        RMSE = math.sqrt(mean_squared_error(df_comparison['actual_open'],
                                              df_comparison['forecast_open']))
        text = f'{code}_RMSE:{RMSE}'
        file1 = open(f"RMSE_{code}.txt","w")#write mode
        file1.write(text)
        file1.close()
    
    except:
        pass
```
